## Prepearing Levada data for merging

```python
import numpy as np
import pandas as pd
```

```python
#get levada-centr data and prepeare it for merging
lev = pd.read_csv("levada.csv")
lev.head()
```

```python
lev['Date'] = pd.to_datetime(lev["Date"])
lev.info()
```

```python
groupLev = lev.groupby([lev.Date.dt.year, lev.Date.dt.month]).mean() #may 2016 has two results in levada.csv
groupLev
```

```python
#getting dates as columns
groupLev.index.rename(['year','month'], inplace = True)
groupLev.reset_index(inplace=True)
groupLev
```

## Aggregating main data and prepearing it for merging

```python
res = pd.read_csv('RESULT.csv')
res.head()
```

```python
res["date"] = pd.to_datetime(res["date"])
res.info()
```

```python
groupRes = res.groupby([res.date.dt.year, res.date.dt.month]).sum() #getting count of each article type each month
```

```python
groupRes.index.rename(['year','month'], inplace = True)
groupRes.reset_index(inplace=True)
groupRes
```

## Merging data and creating correlation table

```python
#merge data
dataLev = pd.merge(groupLev, groupRes, on = ['year','month'])
```

```python
dataLev.head()
```

```python
#making columns with percentages of articles of each type
dataLev['NegPercentage'] = dataLev['Negative']/(dataLev['Negative']+dataLev['Positive']+ dataLev['Neutral'])
dataLev['PosPercentage'] = dataLev['Positive']/(dataLev['Negative']+dataLev['Positive']+ dataLev['Neutral'])
dataLev['NeuPercentage'] = dataLev['Neutral']/(dataLev['Negative']+dataLev['Positive']+ dataLev['Neutral'])
dataLev = dataLev.drop(index = [0,1]) #those two observations are going to be severe outliers
dataLev.head()
```

```python
dataLev.to_csv("fullLevadaPrepeared.csv")
```

```python
#getting correlation table with categories united
dataLev["GoodAll"] = dataLev["Good"] + dataLev["Very Good"]
dataLev["BadAll"] = dataLev["Bad"] + dataLev["Very Bad"]
dataLev
```

```python
corTable = dataLev[dataLev.columns[2:]].corr()
finalcor = corTable[["NegPercentage" ,"PosPercentage" ,"NeuPercentage"]].loc[['Hard to answer', 'GoodAll', 'BadAll'],:]
```

```python
finalcor.to_csv("corTableUnitedCategories.csv")
```
